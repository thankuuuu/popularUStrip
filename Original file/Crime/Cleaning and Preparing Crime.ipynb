```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('estimated_crimes_1979_2018.csv')
df.rename(columns={'state_abbr':'State'}, inplace=True)
df = df.fillna(0)
df = df.drop(['caveats','rape_legacy'],axis=1)
df.head()
```

```python
# Change State=0 to all
df["State"]= df["State"].replace(0, "All")
```

```python
#Add Total Crime
df['Total']= df.iloc[:, -9:].sum(axis=1)
df.head()
```

```python
#Subsetting each year
Data_2016 = df.loc[(df.year == 2016),]
Data_2017 = df.loc[(df.year == 2017),]
Data_2018 = df.loc[(df.year == 2018),]
```

```python
Data_2016.head()
```

```python
Data = pd.concat([Data_2016,Data_2017,Data_2018], axis=0)
Data.shape
```

```python
# Export as csv file
Data.to_csv('USCrime.csv', index = False)
```
