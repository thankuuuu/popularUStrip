```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('us-dollar-index-historical-chart.csv', header = 8)
df.head()
```

```python
# Extracing Month, Year and add the season
month = []
year = []
season = []
for i in range (len(df.index)):
        year.append((df.loc[i,'date'][0:4]))
        month.append((df.loc[i,'date'][5:7]))  
        if df.loc[i,'date'][5:7] in ['03','04','05'] :
            season.append('Spring')
        elif df.loc[i,'date'][5:7] in ['06','07','08'] :
            season.append('Summer')
        elif df.loc[i,'date'][5:7] in ['09','10','11'] :
            season.append('Fall')
        elif df.loc[i,'date'][5:7] in ['12','01','02'] :
            season.append('Winter') 
    
```

```python
join = {'Month':month,'Year':year,'Season':season}
df2 = pd.DataFrame(join)
df2.shape
```

```python
USIndex = pd.concat([df,df2], axis=1)
USIndex.shape
```

```python
Data_2016 = USIndex.loc[(USIndex.Year == '2016'),]
Data_2017 = USIndex.loc[(USIndex.Year == '2017'),]
Data_2018 = USIndex.loc[(USIndex.Year == '2018'),]
Data_2019 = USIndex.loc[(USIndex.Year == '2019'),]
Data = pd.concat([Data_2016,Data_2017,Data_2018,Data_2019], ignore_index=True)
Data.shape
```

```python
Data1 = Data.drop(['date'],axis =1)
Data1.head(10)
```

```python
Data1.rename(columns={' value':'USIndex'}, inplace=True)
Data1
```

```python
# Export as csv file
Data.to_csv('USIndex.csv', index = False)
```
